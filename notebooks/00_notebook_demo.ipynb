# Notebook demo — state, cells, and why order matters

You will **intentionally** break this notebook by running cells out of order — then recover using **Restart Kernel → Run All**.


```python
x = 10
print('x is', x)
```

```python
# Run this cell BEFORE the next one.

def bump():
    global x
    x = x + 1

bump()
print('after bump, x is', x)
```

```python
# Now intentionally run this cell FIRST (before the earlier ones) and see what happens.
print('x squared is', x**2)
```

- How did **Restart Kernel → Run All** fix it?
